## Data Preparation by Python 

```scala
%python
# Filter useful columns
useless_column = ['dob','sofifa_id','player_url','long_name','body_type','real_face','nation_position','loaned_from','nation_jersey_number']
fifa_21 = df.drop(useless_column, axis = 1)
```

```scala
%python
# Create a new columns to calculate BMI
fifa_21['BMI'] = fifa_21 ['weight_kg'] / (fifa_21['height_cm'] / 100) ** 2
```

```scala
%python
# checking missing values 
fifa_21.isnull().any().any()
```

```scala
%python
# Fill the missing value to 0 
fifa_21 = fifa_21.fillna(0)
```

```scala
%python
# checking missing values again
fifa_21.isnull().any().any()
```

## Data Discovery with SQL

```scala
%python.sql
select preferred_foot, count(short_name) as player_count from fifa_21 group by preferred_foot
```

```scala
%python.sql
select age, avg(potential) as avg_potential from fifa_21 where age <=40 group by age order by age asc 
```

```scala
%python.sql
select age, preferred_foot, round(avg(overall),1) as avg_rating from fifa_21 where (age >= 20 and age <=30) group by age,preferred_foot
```

```scala
%python.sql
select nationality, count(short_name) as players_num from (select short_name, nationality, overall from fifa_21 order by overall desc limit 50) group by nationality order by players_num desc 
```
